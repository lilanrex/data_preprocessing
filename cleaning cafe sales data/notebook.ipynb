```python
import pandas as pd
```

```python
df = pd.read_csv('dirty_cafe_sales.csv')
df
```

```python
df.dtypes # we use this to check the data types in which the columns are stored in, in this case it prints object mostly means string or mixed data type.
```

```python
df.info()
```

```python
'''So in order to clean our data, we have to first convert the 'quantity' from object to int64 dtype,
then price and total spent to float64 dtype,then transaction date to datetime64[ns] '''
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce') 
#this converts the quantity column to a numeric dtype(int or float), and the coerce simply means if it encounters 
#something it can't convert, it should make it NaN
```

```python
df.info()
```

```python
df.head()
```

```python
df['Price Per Unit']  = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df.head()
```

```python
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors= 'coerce')
df.head()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors ='coerce')
df.head()
```

```python
df.info()
```

```python
'''so we have changed the dtype to numeric type
now we have to sort the unknowns in item, payment method and location'''
import numpy as np
columns_with_unknown = ['Item', 'Payment Method', 'Location']
for col in columns_with_unknown:
    if col in df.columns:
        unknown_count = df[col][df[col] == 'UNKNOWN'].count()
        if unknown_count > 0:
            df[col] = df[col].replace('UNKNOWN', np.nan)
        else:
            print('zero unknowns found')

df.head()           
```

```python
'''now that we've made corrections to the dtypes and changed all unknowns and type coercions to NaN, we need to 
inspect the NaN more'''

print(df.isnull().sum())
```

```python
df.info()
```

```python
'''since we don't have data to fill in the NaNs we have to simply drop it '''
df.isna().sum()
```

```python
new_df = df.dropna()
```
